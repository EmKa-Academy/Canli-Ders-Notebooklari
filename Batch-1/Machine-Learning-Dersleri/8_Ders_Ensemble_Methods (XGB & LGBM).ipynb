## Outlier Analysis

### IQR

```python
# Boxplot IQR analizini otomatik yapar ve grselletirir.

for col in df.columns[:-1]:
  sns.boxplot(df[col] , whis= 2.5)
  plt.title(col)
  plt.show()
```

```python
df['Age (day)']
```

```python
# Manual IQR
q1 = df['Age (day)'].quantile(0.25)
q3 = df['Age (day)'].quantile(0.75)

IQR = q3 - q1

up_limit = q3 + 1.5 * IQR
low_limit = q1 - 1.5 * IQR

outliers = df[(df['Age (day)'] > up_limit) | (df['Age (day)'] < low_limit) ]

outliers
```

```python
# IQR Function
def iqr_analysis(df , col , iqr_rate= 1.5):
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)

  IQR = q3 - q1

  up_limit = q3 + iqr_rate * IQR
  low_limit = q1 - iqr_rate * IQR

  outliers = df[(df[col] > up_limit) | (df[col] < low_limit)]

  return outliers
```

```python
iqr_analysis(df , df.columns[1] , 1.5 )
```

# Modelling

### Train | Test Split

```python
X = df.drop(target_col , axis = 1)
y = df[target_col]
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 34)
```

```python
print(f"X shape {X.shape}")
print(f"X_train shape {X_train.shape}")
print(f"X_test shape {X_test.shape}")
```

# Model Time

```python
# Base XGB , Base LGBM , Tuned XGB , Tuned LGBM
```

# Base XGB

```python
from xgboost import XGBRegressor
```

```python
base_xgb = XGBRegressor()

base_xgb.fit(X_train , y_train)
```

```python
y_pred = base_xgb.predict(X_test)
y_pred
```

```python
np.array(y_test)
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
mean_absolute_error(y_test , y_pred)
```

```python
mean_squared_error(y_test , y_pred) # mse ok anlaml deil.
```

```python
mean_squared_error(y_test , y_pred) ** 0.5 # RMSE, MAE benzeri bir metrik gibi dnlebilir ancak byk hatalar ok daha byk cezalandrr.
```

```python
r2_score(y_test , y_pred)
```

```python
def eval_metrics(y_test , y_pred):
  mae = mean_absolute_error(y_test , y_pred)
  mse = mean_squared_error(y_test , y_pred)
  rmse = mean_squared_error(y_test , y_pred) ** 0.5
  r2 = r2_score(y_test , y_pred)

  print(f"""
  MAE : {mae}
  MSE : {mse}
  RMSE : {rmse}
  R2 : {r2}
  """)
```

```python
eval_metrics(y_test , y_pred)
```

# Tuned XGB

### RandomSearch

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
vanilla_xgb = XGBRegressor()

param_dist = {
    'n_estimators': np.arange(100,2000,100), # [100,200,300,400,....,2000]
    'max_depth' : np.arange(2, 12 , 1),
    'learning_rate': np.arange(0.01 , 1 , 0.05),
    'subsample': [0.5 , 0.6 , 0.7, 0.8 , 0.9 , 1]
}

random_model = RandomizedSearchCV(vanilla_xgb , param_dist , n_iter = 30 , cv = 5 , scoring = 'neg_mean_squared_error')

random_model.fit(X_train, y_train)
```

```python
random_model.best_params_
```

```python
tuned_xgb = random_model.best_estimator_
```

```python
y_pred = tuned_xgb.predict(X_test)
eval_metrics(y_test , y_pred)
```

## Tuned LGBM

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
vanilla_lgb = LGBMRegressor()

params = {
    'n_estimators' : [100, 300 , 600 , 1000],
    'max_depth' : [3, 5, 7] ,
    'learning_rate' : [0.01 , 0.07 , 0.1],
    'subsample' : [0.5 , 0.8 , 1]
}

grid_model = GridSearchCV(vanilla_lgb , params , scoring= 'neg_mean_squared_error' , cv = 5 )
grid_model.fit(X_train , y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
tuned_lgb = grid_model.best_estimator_
```

```python
y_pred = tuned_lgb.predict(X_test)
eval_metrics(y_test , y_pred)
```
