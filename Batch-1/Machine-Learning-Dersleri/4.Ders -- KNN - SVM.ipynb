# Boyut İndirgeme Yöntemleriyle Görselleştirme (t-SNE , PCA , LDA)

* PCA (Principal Component Analysis), verilerin varyansını koruyarak boyut indirgeme yapar ve değişkenlikleri açıklamak için kullanılır.
* LDA (Linear Discriminant Analysis), sınıflar arasındaki ayrımı arttırmak için boyut indirgeme yapar ve sınıflandırma problemlerinde kullanılabilir.
* t-SNE (t-Distributed Stochastic Neighbor Embedding), verilerin benzerliklerini koruyarak boyut indirgeme yapar ve verilerin kümelenmelerini ve yapısını görselleştirmek için kullanılır.

## t-SNE

```python
# X = df.drop('Room' , axis = 1)
# y = df['Room']

features = df.drop('Room' , axis = 1)
target = df['Room']
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2 , random_state= 53)
tsne_result = tsne.fit_transform(features)
```

```python
tsne_result # Sonuçlar
```

```python
tsne_df = pd.DataFrame(data = tsne_result, columns= ['t-SNE Component 1' , 't-SNE Component 2'])
tsne_df['Room'] = target
tsne_df
```

```python
sns.scatterplot(x='t-SNE Component 1', y='t-SNE Component 2', hue='Room', data=tsne_df, palette='colorblind')
plt.title('t-SNE Görselleştirme')
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state= 53)
pca_result = pca.fit_transform(features)
```

```python
pca_df = pd.DataFrame(data = pca_result, columns = ['PCA Component 1' , 'PCA Component 2'])
pca_df['Room'] = target
pca_df
```

```python
sns.scatterplot(x='PCA Component 1', y='PCA Component 2', hue='Room', data=pca_df, palette='colorblind')
plt.title('PCA Görselleştirme')
plt.show()
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components= 2)
lda_result = lda.fit_transform(features, target)
```

```python
lda_df = pd.DataFrame(lda_result, columns= ['LDA Component 1' , 'LDA Component 2'])
lda_df['Room'] = target
```

```python
sns.scatterplot(x='LDA Component 1', y='LDA Component 2', hue='Room', data=lda_df, palette='colorblind')
plt.title('LDA Görselleştirme')
plt.show()
```

# Model Time

### Train | Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features , target , test_size = 0.2 , random_state= 53)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

# Model Time

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors= 5)
```

```python
knn_model.fit(X_train, y_train)
```

```python
knn_y_pred = knn_model.predict(X_test)
```

```python
knn_y_pred
```

```python
np.array(y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix , classification_report

accuracy_score(y_test, knn_y_pred)
```

```python
print(classification_report(y_test, knn_y_pred))
```

```python
print(confusion_matrix(y_test, knn_y_pred))
```

```python
knn_model.predict([[-15,-17,-44,-21,-39,-23,-99]])
```

#### CV ile KNN Accuracy

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(knn_model , features, target , cv= 5)

cv_scores
```

```python
cv_scores.mean() # Daha güvenilir bir sonuç
```

```python
print(f"KFold CrossValidation yaparak test ettiğimiz KNN modelinin Accuracy : {cv_scores.mean()}")
```

### SVC

```python
from sklearn.svm import SVC

svc_model = SVC(kernel= 'rbf' , C= 300 , random_state= 53)
```

```python
svc_model.fit(X_train,y_train)

svc_y_pred = svc_model.predict(X_test)
```
