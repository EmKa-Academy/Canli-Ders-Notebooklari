# DataFrame Basics

```python
# DF oluşturma.
# DF'in içerisinde bir ya da birden fazla column'u(sütunu) seçmek.
# DF'in içerisinde bir ya da birden fazla row'u(satırı) seçmek.
# DF'in içerisine bir column ya da row ekleme.
```

```python
import numpy as np
import pandas as pd
```

```python
# pd.DataFrame() fonksiyonuyla basitçe df oluşturabiliriz.
```

```python
my_data = np.random.randint(0,101 , size= (5,4))
my_data
```

```python
pd.DataFrame(my_data)
```

```python
my_idx = ['TR' , 'BR' , 'USA' , 'RU']
my_cols = ['Jan' , 'Feb' , 'Mar']
my_data = np.random.randint(0,101, size= (4,3))
```

```python
df = pd.DataFrame(data = my_data , index = my_idx , columns = my_cols)
df
```

```python
df.index = ['ali' , 'mehmet' , 'burak' , 'aslı']
```

```python
df.columns = ['1-Not' , '2-Not' , '3-Not']
```

```python
df
```

```python
df.info() # .info() metoduyla DF'in içindeki bütün columnların temel bilgilerini görebiliriz.
```
